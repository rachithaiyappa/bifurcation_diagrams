```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
c = 0.60
N = 30
x1 = np.linspace(0,c/N,20)
x2 = np.linspace(c/N,2*c/N,20)
y1 = np.linspace(0,0,20)
y2 = np.linspace(0,0,20)
y3 = np.linspace(1,1,20)
y4 = np.linspace(-1,-1,20)
plt.title('Pairwise stochastic system')
plt.plot(x1,y1,'r.',label='m1 unstable')
plt.plot(x2,y2,'r',label='m1 stable')
plt.plot(x1,y3,'b')
plt.plot(x1,-y3,'b',label='m2 stable')
plt.plot(x2,y4,'b.',label='m2 ustable')
plt.plot(x2,-y4,'b.')
plt.annotate(r'$\frac{c}{N}$',(c/N,0),(c/N,-0.1))
plt.legend(loc='upper right')
plt.xlabel('s')
plt.ylabel('m',rotation=0)
plt.ylim(-1.1,1.1)
plt.show()
```

```python
x1 = np.linspace(0,0.2,20)
y1 = np.linspace(0,0,20)
plt.title('Pairwise deterministic system')
plt.plot(x1,y1,'r',label='m1 stable')
plt.legend(loc='upper right')
plt.xlabel('s')
plt.ylabel('m',rotation=0)
plt.ylim(-1,1)
plt.show()
```

```python
h = 0.30
x1 = np.linspace(0,h/4,20)
x2 = np.linspace(h/4,2*h/4,20)
y = np.sqrt(1-4*x1/h)
plt.title('Higher order deterministic system')
plt.plot(x1,y,'b',label='m1 stable')
plt.plot(x1,-y,'b')
plt.plot(x1,y*0,'r.',label='m2 unstable')
plt.plot(x2,y*0,'r',label='m2 stable')
plt.plot(h/4,0,'+')
plt.annotate('h/4',(h/4,0),(h/4,-0.1))
plt.annotate(r'$\sqrt{1-4s/h}$',(h/8,np.sqrt(1-4*h/8/h)),(h/8,np.sqrt(1-4*h/8/h)))
plt.legend(loc='upper right')
plt.xlabel('s')
plt.ylabel('m',rotation=0)
plt.show()
```

```python
h = 1
c = 1
N = 1
length = 20
x1 = np.linspace(0,(h/4)+(h/(2*N))+(c/N),length)
x2 = np.linspace((h/4)+(h/(2*N))+(c/N),2*((h/4)+(h/(2*N))+(c/N)),length)
y = np.sqrt(1-(4*x1/h)+((4*c)/(h*N))+(2/N))
plt.title('Higher order stochastic system')
plt.plot(x1,y,'b',label='m1')
plt.plot(x1,-y,'b')
plt.plot(x1,y*0,'r.',label='m2 unstable')
plt.plot(x2,y*0,'r',label='m2 stable')
plt.annotate(r'$\frac{h}{4}+\frac{h}{2N}+\frac{c}{N}$',((h/4)+(h/(2*N))+(c/N),0),((h/4)+(h/(2*N))+(c/N),-0.4))
plt.annotate(r'$\sqrt{1-\frac{4s}{h}+\frac{4c}{hN}+\frac{2}{N}}$',(0.5*((h/4)+(h/(2*N))+(c/N)),np.sqrt(1-((4*(0.5*((h/4)+(h/(2*N))+(c/N))))/h)+((4*c)/(h*N))+(2/N))))
plt.legend(loc='upper right')
plt.show()
```

```python
def f(h):
    plt.figure()
    x1 = np.linspace(0,h/4,20)
    x2 = np.linspace(h/4,2,20)
    y = np.sqrt(1-4*x1/h)
    plt.title('Higher order deterministic system')
    plt.plot(x1,y,'b',label='m1 stable')
    plt.plot(x1,-y,'b')
    plt.plot(x1,y*0,'r.',label='m2 unstable')
    plt.plot(x2,y*0,'r',label='m2 stable')
    plt.annotate('h/4',(h/4,0),(h/4,-0.1))
    plt.annotate(r'$\sqrt{1-4s/h}$',(h/8,np.sqrt(1-4*h/8/h)),(h/8,np.sqrt(1-4*h/8/h)))
    plt.legend(loc='upper right')
    plt.ylim(-1,1)
    plt.xlim(0,2)
    plt.show()
    
    
interact(f,h=(0,10));
```
